```python
from sklearn.impute import SimpleImputer 
```

```python
# we have to create an object from the Simple Imputer 
# hence we have to create a var 
sp = SimpleImputer()
```

```python
#age = df["Age"].reshape(-1,1)
```

```python
sp.fit_transform(df[["Age"]]) # we cannot use implace to replace the new age data hence we create a vriable and add there
```

```python
df["Age"] = sp.fit_transform(df[["Age"]])
```

```python
df.isnull().sum() # implaced with the new age 
```

```python
df.info()
```

```python
sp1 = SimpleImputer(strategy = "most_frequent")
```

```python
df["Embarked"] = sp1.fit_transform(df[["Embarked"]])
```

```python
df.isnull().sum()
```

```python
df = df.drop(["Cabin"], axis = 1)
```
